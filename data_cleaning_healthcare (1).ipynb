```python
#importing files
import pandas as pd
import numpy as np
```

```python
#load csv file
ds=pd.read_csv('messy_healthcare_data_2.csv')
ds
```

```python
#serching duplicates in file
ds.duplicated().sum()
```

```python
#drop null values of Patient_id columns
ds=ds.dropna(subset=['Patient_ID'])
ds
```

```python
#get info about dataset
ds.info()
```

```python
#handle full_name column
ds['Full_Name'].replace({'12345':'Unknown'},inplace=True)
ds
```

```python
#handle DOB column
ds['DOB']=pd.to_datetime(ds['DOB'],errors='coerce')
ds
```

```python
#handling height_cm column
ds['Height_cm'].replace({'one seventy':'170','?':'NaN'},inplace=True)
ds
```

```python
#handling weight_kg
ds['Weight_kg'].replace({'eighty':'80'},inplace=True)
ds
```

```python
#handling smoker_status
ds['Smoker_Status'].replace({'Y':'Yes','N':'No','Unknown':'NaN'},inplace=True)
```

```python
ds
```

```python
#handling allergies column
ds['Allergies'].replace({'123':'NaN'},inplace=True)
ds
```

```python
#handling medications coulmn
ds['Medications'].replace({'12345':'NaN'},inplace=True)
ds
```

```python
#handling Visit_Date column
ds['Visit_Date']=pd.to_datetime(ds['Visit_Date'],errors='coerce')
```
